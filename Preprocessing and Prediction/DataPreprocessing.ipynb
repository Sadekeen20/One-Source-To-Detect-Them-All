##Removing the Outliers##

```
#find Q1, Q3, and interquartile range for each column
Q1 = gender.quantile(q=.25)
Q3 = gender.quantile(q=.75)
IQR = abs(Q1-Q3)

#only keep rows in dataframe that have values within 1.5*IQR of Q1 and Q3
gender = gender[~((gender < (Q1-1.5*IQR)) & (gender > (Q3+1.5*IQR))).any(axis=1)]

#find how many rows are left in the dataframe 
gender.shape
```

```
#find Q1, Q3, and interquartile range for each column
Q1 = age.quantile(q=.25)
Q3 = age.quantile(q=.75)
IQR = abs(Q1-Q3)

#only keep rows in dataframe that have values within 1.5*IQR of Q1 and Q3
age = age[~((age < (Q1-1.5*IQR)) & (age > (Q3+1.5*IQR))).any(axis=1)]

#find how many rows are left in the dataframe 
age.shape
```

```
#find Q1, Q3, and interquartile range for each column
Q1 = emotion.quantile(q=.25)
Q3 = emotion.quantile(q=.75)
IQR = abs(Q1-Q3)

#only keep rows in dataframe that have values within 1.5*IQR of Q1 and Q3
emotion = emotion[~((emotion < (Q1-1.5*IQR)) & (emotion > (Q3+1.5*IQR))).any(axis=1)]

#find how many rows are left in the dataframe 
emotion.shape
```

##SMOTE on Gender Dataset##

```
from imblearn.over_sampling import SMOTE
df=gender
df=df[['meanfreq','sd','median','Q25','Q75','IQR','skew','kurt','sp.ent','sfm','mode','centroid','meanfun','minfun','maxfun','meandom','mindom','maxdom','dfrange','modindx','label']]
seed = 100
k = 1

X = df.loc[:, df.columns != 'label']
y = df.label
sm = SMOTE(sampling_strategy='auto', k_neighbors=1, random_state=seed)
X_res, y_res = sm.fit_resample(X, y)

df = pd.concat([pd.DataFrame(X_res), pd.DataFrame(y_res)], axis=1)
# rename the columns
df.columns = ['meanfreq']+['sd']+['median']+['Q25']+['Q75']+['IQR']+['skew']+['kurt']+['sp.ent']+['sfm']+['mode']+['centroid']+['meanfun']+['minfun']+['maxfun']+['meandom']+['mindom']+['maxdom']+['dfrange']+['modindx']+['label']
df.to_csv('smoted_gender.csv', index=False, encoding='utf-8')
```
